```
import matplotlib.pyplot as plt
x=[3,8]
x1=[4,7]
y=[7,14]
y1=[3,8]
```

```
plt.xlabel("Time")
plt.ylabel("Distance")
plt.plot(x,y,label="Speed of car")
plt.bar(x,y,label="Speed of car1")
plt.plot(x1,y1,label="Speed of Truck")
plt.bar(x1,y1,label="Speed of Truck1")
plt.grid(color='green')
plt.legend()
plt.show()
```
